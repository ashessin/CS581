```python
client = MongoClient('mongodb://localhost:27017/admin')
client.list_database_names()
```

```python
db = client[DB]
rides_collection = db[COLLECTION]

def insert_records(csv_filepath, year, month):
    processed = skipped = 0
    print(csv_filepath)
    for chunck_df in pd.read_csv(csv_filepath, chunksize=10000):
        processed, trips = 0, []
        for index, row in chunck_df.iterrows():
            processed = index + 1
            pick_point = Point(row['pickup_longitude'], row['pickup_latitude'])
            drop_point = Point(row['dropoff_longitude'], row['dropoff_latitude'])
            if not (POLYGON.contains(pick_point) or POLYGON.contains(drop_point)):
                skipped = skipped + 1
                continue
            trip = {
                '_id' : int(year + month + '{:09}'.format(processed)),
                'vendor_id': row['VendorID'],
                'start_timestamp': dt.datetime.strptime(row['tpep_pickup_datetime'], '%Y-%m-%d %H:%M:%S'),
                'end_timestamp': dt.datetime.strptime(row['tpep_dropoff_datetime'], '%Y-%m-%d %H:%M:%S'),
                'pick_location': {
                    'type': "Point", 'coordinates': [ row['pickup_longitude'], row['pickup_latitude'] ] 
                },
                'drop_location': {
                    'type': "Point", 'coordinates': [ row['dropoff_longitude'], row['dropoff_latitude'] ]
                },
                'distance': row['trip_distance'],
                'rate_code_id': row['RateCodeID'],
                'is_store_and_fwd': True if row['store_and_fwd_flag'].upper() == 'Y' else False,
                'passengers': row['passenger_count'],
                'payment_type_id': row['payment_type'],
                'fare': {
                    'base': row['fare_amount'],
                    'tips': row['tip_amount'],
                    'tolls': row['tolls_amount'],
                    'extra': row['extra'],
                    'mta_tax': row['mta_tax'], 
                    'improvement_surcharge': row['improvement_surcharge'],
                    'total': row['total_amount']
                },
                'year': int(year)
            }
            trips.append(trip)
        rides_collection.insert_many(trips)
        print('\r' + f'Number of records processed: \t{processed}', end='')
    print()
    print(f'Number of records inserted: \t{processed - skipped}')
    print(f'Number of records skipped: \t{skipped}')

for i in range(1, 13):
    year, month = '2015', '{:02}'.format(i)
    filepath = f"/src/datasets/yellow_tripdata_{year}-{month}.csv"
    if os.path.exists(filepath):
        insert_records(filepath, year, month)
    else:
        print(f'File {filepath} not found.)
```
